## Improting libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset and making a shallow copy of dataset

```python
dataOriginal = pd.read_csv(r'F:\data\mushrooms.csv')
```

```python
data = dataOriginal.copy()
```

## Understanding Dataset

```python
data.shape
```

## Setting parameter to display all columns

```python
pd.set_option("display.max_columns",None)
```

```python
data.head()
```

## Statistics of data

```python
data.describe()
```

## Information about data

```python
data.info()
```

## Finding unique values of data

```python
data.nunique()
```

### In "veil-type" columns we have only one unique value so we are droping that column.

```python
data.drop('veil-type',axis=1,inplace=True)
```

## Checking total number of null values in every column. 

```python
data.isnull().sum()
```

## We are not having any null value in any of the columns.

## Data Manipulation

The datatype of all columns is object, so for applying model we have to convert datatype to category.

```python
data = data.astype("category")
```

```python
data.info()
```

```python
data.dtypes
```

## Data destribution in previous output columns.

```python
sns.countplot(data=data,x='class')
```

## We are Transforming categorical data to numerical data by Label Encoding.

```python
# IMPORTING LIBRARY FOR LABEL ENCODINGabs
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for column in data.columns:
    data[column] = le.fit_transform(data[column])
```

```python
data.head()
```

## Saperating feature matrix 'x' and Response matrix 'y'.

```python
x = data.drop('class',axis=1)
y = data['class']
```

```python
x.shape
```

```python
y.shape
```

```python
x.head()
```

```python
y.head()
```

## We can take a look on correlation between independent variables.

```python
fig , ax = plt.subplots(figsize = (18,10))
sns.heatmap(x.corr(), cmap="YlGnBu", annot=True)
```

## Feature 'veil-color' is highly correlated with feature 'gill-attachment' so we are dropping 'gill-attachment' feature.

```python
x.drop('gill-attachment',axis=1,inplace=True)
```

```python
fig , ax = plt.subplots(figsize = (18,10))
sns.heatmap(x.corr(), cmap="YlGnBu", annot=True)
```

## Applying Principal Component Analysis (PCA)

```python
# IMPORTING LIBRARY FOR PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
x_pca = pca.fit_transform(x)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
pd.DataFrame(x_pca).head()
```

## Dividing data into training data set and testing data set.

```python
# IMPORTING LIBRARY FOR TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_pca, y, test_size=0.2,random_state=42)
```

```python
x_train.shape , y_train.shape
```

```python
x_test.shape , y_test.shape
```

# Applying Machine Learning Models

## Importing 'accuracy_score' to find accuracy of Machine Learning models.

```python
from sklearn.metrics import accuracy_score
```

## Importing LogisticRegression model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
lr_accuracy_score = accuracy_score(y_test,y_pred_lr)
```

```python
print('Logistic Regression Accuracy :- ',lr_accuracy_score)
```

## Importing RandomForestClassifier model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
rf_accuracy_score = accuracy_score(y_test,y_pred_rf)
```

```python
print('Random Forest Accuracy :- ',rf_accuracy_score)
```

## Importing KNeighborsClassifier model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
knn_accuracy_score = accuracy_score(y_test,y_pred_knn)
```

```python
print('K Neighbors Accuracy :- ',knn_accuracy_score)
```

## Importing GradientBoostingClassifier model

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
```

```python
gb.fit(x_train,y_train)
```

```python
y_pred_gb = gb.predict(x_test)
```

```python
gb_accuracy_score = accuracy_score(y_test,y_pred_gb)
```

```python
print('Gradient Boosting Accuracy :- ',gb_accuracy_score)
```

## Importing DecisionTreeClassifier model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
dt_accuracy_score = accuracy_score(y_test,y_pred_dt)
```

```python
print('Decision Tree Accuracy :- ',dt_accuracy_score)
```

## Importing Support Vector Classifier (SVC) model

```python
from sklearn.svm import SVC
sv = SVC()
```

```python
sv.fit(x_train,y_train)
```

```python
y_pred_sv = sv.predict(x_test)
```

```python
sv_accuracy_score = accuracy_score(y_test,y_pred_sv)
```

```python
print('Support Vector Accuracy :- ',sv_accuracy_score)
```

## Creating Data Frame and Visualize accuracy

```python
acc_matrix = pd.DataFrame({"Model":['LR','RF','KN','GB','DT','SV'],
                           "ACC":[lr_accuracy_score*100,
                                 rf_accuracy_score*100,
                                 knn_accuracy_score*100,
                                 gb_accuracy_score*100,
                                 dt_accuracy_score*100,
                                 sv_accuracy_score*100]})
```

```python
acc_matrix
```

```python
sns.barplot(data=acc_matrix,x='Model',y='ACC',palette="Greens_d")
```
